## 1. Wprowadzenie



Chcąc rozważyć temat modeli epidemiologicznych, należałoby zacząć od wprowadzenia do tematu, który będziemy rozważać w dalszej części pracy oraz krótkiego zarysu przedstawiającego jak rozwijało się modelowanie w epidemiologii na przestrzeni wieków.

Słowo epidemia pochodzi z języka greckiego (epi – nawiedzający, demos – ludzi), a więc można je rozumieć jako wystąpienie na danym obszarze zakażeń lub zachorowań na chorobę zakaźną w liczbie wyraźnie większej niż we wcześniejszym okresie. Epidemia chorób zakaźnych występuje lub bierze swój początek zwykle na terenie, na którym zachorowania występowały endemicznie (tylko na danym terenie), może jednak wystąpić na obszarze dotychczas wolnym od tej choroby. Źródłem zakażenia jest najczęściej chory człowiek lub nosiciel zarazków, chore zwierzę lub produkty pochodzenia zwierzęcego. Zarazki mogą się przenosić przez bezpośredni kontakt z chorym, przez powietrze, wodę, pokarmy lub np. za pośrednictwem niektórych stawonogów, lub pierwotniaków. To czy dojdzie do zakażenia, w dużej mierze zależy od zdolności do wywołania choroby danego patogenu, a także od odporności własnej pacjenta, dawki i wirulencji czynnika zakaźnego oraz wpływu dodatkowych czynników.

Jako najbardziej znane epidemie w historii uważa się:

Choroby te spustoszyły większą część średniowiecznego świata, dlatego ludzie zaczęli reagować. Chcąc uchronić się od kolejnej zarazy, zaczęli szukać sposobów, na unikniecie epidemii w przyszłości. Ówcześni mieszczanie zaczęli stosować odpowiednie diety (np. na południu Europy zalecano jeść dużo cytrusów, a na północy – kapusty). Zwiększono nacisk na higienę. Zalecano myć włosy w wywarze z szałwii, piołunu i róży, nie kąpać się z innymi i nie golić się nie swoją brzytwą. Naukowcy rozpoczęli prace nad szczepionkami, które miały wyeliminować choroby. 

Matematycy również nie pozostali obojętni. Pierwszy znany opis matematyczny procesu rozprzestrzeniania się epidemii został wykonany przez Bernoullego w 1760 roku. Mimo że, od tych czasów nauki medyczne bardzo się rozwinęły, podobnie zresztą jak różne matematyczne metody opisu procesów rozprzestrzeniania się, do dzisiejszego dnia równania różniczkowe stanowią podstawowe narzędzie badawcze do modelowania dynamiki chorób zakaźnych. Takim ponadczasowym modelem rozwijania się epidemii, opartym na układzie równań różniczkowych zwyczajnych i mającym ogromny wpływ na rozwój modelowania matematycznego w tym zakresie był model SIR zaproponowany przez Kermacka i McKendricka na początku ubiegłego wieku.[2]

Modele epidemiologiczne pozwalają na przewidywanie rozwoju epidemii. Umożliwiają one określenie czasu jej trwania, szybkości jej rozwoju czy też aktualnej liczby osobników zarażonych i ozdrowiałych.

## 2. Podstawowy model SIR

SIR to model epidemii, który dzieli izolowaną populację na trzy grupy:
- S (ang. suspectible) - osobniki podatne na chorobę,
- I (ang. infected) - osobniki chorujące i roznoszące infekcję,
- R (ang. recovered) - osobniki ozdrowiałe, które nabyły odporność.
    (w tym osobniki, które umarły na skutek przebytej infekcji (ang. removed)).

Model SIR zakłada, że:

- osobniki z grupy R są usunięte z populacji poprzez śmierć, izolację bądź są uodpornione na powtórne zarażenie wskutek przebytej choroby;
- liczebność badanej populacji nie zmienia się w czasie;
- rozprzestrzenianie się epidemii polega na tym, że osoby chore zarażają osoby zdrowe, przekazując im patogen (wirus lub bakteria);
- chore osobniki po wyzdrowieniu stają się odporne, tzn. nie są podatne na infekcję wywołaną tym samym patogenem;
- okres inkubacji jest na tyle krótki, że można go zaniedbać — osobnik podatny, który się zaraził, zaczyna chorować natychmiast i jest zdolnym zarażać pozostałe osobniki;
- jednostki zmarłe również uważa się za odporne i w dalszym ciągu się je zlicza;
- każda para osobników ma równe prawdopodobieństwo wzajemnego kontaktu.

Przejścia między wymienionymi grupami opisuje schemat:
 ![](schemat2.png " Model SIR dla chorób powodujących odporność")

Osobnik zdrowy (podatny, S) mający kontakt z osobnikiem chorym (zainfekowanym, I), z pewnym prawdopodobieństwem, zależnym m.in. od rodzaju patogenu, może zostać zarażony wirusem bądź bakterią. Po pewnym czasie jednak, taki chory osobnik zdrowieje (zmienienia stan na R), uzyskując przy tym odporność na daną chorobę.

Przyjmując, że wielkości (liczebności) poszczególnych grup osobników: podatnych, zainfekowanych i ozdrowiałych są odpowiednio równe: S, I oraz R możemy zapisać równania Kermacka-McKendricka. [2]

Układ równań opisujący rozprzestrzenianie się epidemii w czasie:

$$
\left\{ \begin{array}{ll}
\begin{equation}
\begin{split}
\frac{dS}{dt} & = & -rS(t)I(t),\\
\frac{dI}{dt} & = & rS(t)I(t)-aI(t),\\
\frac{dR}{dt} & = & aI(t),
\end{split}
\end{equation}
\tag{1}
\end{array} \right.
$$

gdzie: 

r - tempo rozwoju infekcji;

a - tempo zdrowienia;

$\begin{equation}
r,a >0.
\end{equation}$

Uzasadnienie powyższych równań z układu (1):


- (Równanie pierwsze) Liczba osobników w grupie podatnych S maleje (pochodna jest ujemna), ponieważ mogą zachorować. Tempo ubytku osobników S zależy od liczby osób chorych I, ponieważ im więcej jest nosicieli choroby, tym więcej ludzi może być zainfekowanych.

- (Równanie drugie) Liczba osób chorych I zmienia się z dwóch powodów.
  - I wzrasta (wyraz po prawej stronie równania jest dodatni), ponieważ nowe osoby są infekowane.
  - I maleje (wyraz po prawej stronie jest ujemny), ponieważ chorzy zdrowieją.
  

- (Równanie trzecie) Osób ozdrowiałych R (lub martwych) przybywa (pochodna jest dodatnia). [3]

Z założeń wiemy, że populacja jest izolowana, więc całkowita liczba wszystkich osobników jest stała:

<h4><center>
$\begin{equation}
S(t)+I(t)+R(t)=N_0=const.
\tag{2}
\end{equation}$
</center></h4>

Wynika to również z układu równań (1). Gdybyśmy dodali do siebie wszystkie trzy równania, otrzymalibyśmy:

<h4> <center>
$\begin{equation}
\frac{dS}{dt} + \frac{dI}{dt} + \frac{dR}{dt}=0.
\end{equation}$
</center></h4> 

Sensowne dane początkowe dla tego modelu epidemiologicznego, które przyjmiemy to:

$$
\left\{ \begin{array}{ll}
\begin{equation}
\begin{split}
S(0) & = &   S_0 > 0,\\
I(0) & = &  I_0 > 0,\\
R(0) & = &  0. 
\end{split}
\end{equation}
\end{array} \right.
$$

## 3. Badanie stabilności układu

W układzie równań (1) mamy trzy równania i dwie zmienne niezależne, zatem do badania stabilności całego układu wystarczy wziąć dwa z nich. 

Dwa pierwsze równania z układu (1) dają nam:

$$
\left\{ \begin{array}{ll}
\begin{equation}
\begin{split}
\frac{dS}{dt} & = & -rSI,\\
\frac{dI}{dt} & = & rSI-aI.\\
\end{split}
\end{equation}
\tag{3}
\end{array} \right.
$$

Przyrównujemy je do zera, aby znaleźć współrzędne punktów krytycznych układu.

$$
\left\{ \begin{array}{ll}
\begin{equation}
\begin{split}
0 & = & -rIS   ,\\
0 & = & I(rS-a),
\end{split}
\end{equation}
\tag{4}
\end{array} \right.
$$

Rozwiązując układ (4), otrzymujemy jeden punkt stacjonarny, którego współrzędne to:
<h4><center>
   $ I = 0$ , $ S = \frac{a}{r}.$ 
</center></h4>


Do badania stabilności układu potrzebna jest macierz Jacobiego (czyli  macierz zbudowana z pochodnych cząstkowych (pierwszego rzędu) funkcji, której składowymi są funkcje rzeczywiste) dla układu równań (3), gdzie jako $F(S,I)$ będziemy rozumieć równanie pierwsze, a jako $G(S,I)$ równanie drugie:

<h4><center>
$\begin{equation*}
J(S,I) = 
\begin{bmatrix}
\frac{F(S,I)}{dS} & \frac{F(S,I)}{dI} \\
 \frac{G(S,I)}{dS} & \frac{G(S,I)}{dI}\\
 \tag{5}
 \end{bmatrix}
 \end{equation*}$
</center></h4>


Po obliczeniu pochodnych i podstawieniu do macierzy (5), otrzymujemy:
<h4><center>
$\begin{equation*}
J(S,I) = 
\begin{bmatrix}
-rI & -rS \\
 rI & rS-a \\
 \tag{6}
 \end{bmatrix}
 \end{equation*}$
</center></h4>

W punkcie krytycznym macierz (6) ma postać:
    <h4><center>
$\begin{equation}
J\left(\frac{a}{r},0\right) = 
\begin{bmatrix}
0 & -a \\
0 & 0 \\
 \end{bmatrix}
 \end{equation}$
</center></h4>

Wyliczamy jej wartości własne w punkcie $ \left (\frac{a}{r},0\right)$ przyrównując wyznacznik macierzy do zera:

<h4><center>
$\begin{equation*}
0 = 
\begin{vmatrix}
-\lambda & -a \\
 0 & -\lambda \\
 \end{vmatrix}
= \lambda^{2} - 0 = \lambda^{2}
 \end{equation*}$
</center></h4>
Zatem $\begin{equation}
\lambda_{1} = \lambda_{2}=0.
 \end{equation}$

Opierając się na twierdzeniu Hartman'a - Grobman'a, które mówi, że jeżeli układ równań różniczkowych $x'=f(x)$ posiada linearyzację w pobliżu punktu krytycznego i wszystkie wartości własne macierzy Jacobiego są niezerowe, to istnieje takie otoczenie punktu krytycznego, w którym rozwiązania nieliniowego równania są równoważne jego linearyzacji. Możemy zauważyć, że w momencie otrzymania wszystkich zerowych wartości własnych, rozwiązanie układu (3) nie będzie utożsamiane z jego linearyzacją.
Stąd wniosek, że liniowa analiza stabilności układu nie może zostać przeprowadzona metodami poznanymi na tym kursie. Należy więc zbadać zachowanie się modelu w inny sposób. My zastosujemy numeryczną analizę przy użyciu metody Rungego - Kutty.


## 4. Algorytm Rungego-Kutty 

Do numerycznego rozwiązania badanych równań modelu SIR użyjemy algorytmu Rungego-Kutty 4 rzędu, który jest uważany za najlepszą metodą przybliżonego rozwiązania równań różniczkowych zwyczajnych oraz ich układów, nadającą się dla elektronicznych maszyn cyfrowych. Polega ona na znalezienia wartości funkcji $y(t)$ w skończonej liczbie punktów czasowych. W najprostszym przypadku (do którego się tutaj ograniczymy) zakładamy, że punkty te są od siebie równo oddalone, a odległość między nimi nazywamy krokiem czasowym i oznaczamy literą $h$. Rząd naszej metody to 4, czyli w każdym kroku czasowym czterokrotnie wywołujemy funkcję $f$.


Mamy równanie postaci $ y' = f(x,y)$ oraz warunek początkowy $y (x_{0}) = y_{0} $. Chcemy znaleźć kolejne wartości y. Przyjmując dowolne h otrzymujemy:

<h4><center>
$\begin{equation}
\begin{split}
    y_{n+1} & = & y_{n} + \Delta y_{n} \\
    \Delta y_{n} & = & \frac{1}{6} ( k_{1}+k_{2}+k_{3}+k_{4})
\end{split}
\end{equation}$
</center></h4>

gdzie:
<h4><center>
$\begin{equation}
\begin{split}
    k_{1} & = &  hf(x_{n},y_{n}), \\
    k_{2} & = &  hf(x_{n}+\frac{h}{2},y_{n}+ \frac{1}{2}k_{1}), \\
    k_{3} & = &  hf(x_{n}+\frac{h}{2},y_{n}+ \frac{1}{2}k_{2}), \\
    k_{4} & = &  hf(x_{n}+h,y_{n}+ k_{3}). \\
\end{split}
\end{equation}$
</center></h4>

Na podstawie powyższej metody napisany został kod składający się z dwóch funkcji. Pierwsza z nich nazwana SIR opisuje układ równań Kermacka-McKendricka. Jako argumenty przyjmuje wektor czasów oraz macierz wartości funkcji. Druga z nich opisuje algorytm Runge-Kutty 4 rzędu, jako argumenty przyjmując: warunek początkowy, przedział czasowy, w którym ma się wyliczać oraz krok.[4]

```python
import numpy as np
import matplotlib.pyplot as plt


def SIR(t,y):
    S=y[0] #wartości odpowiadające liczbie podatnych na chorobę przechowywane są w pierwszej kolumnie
    I=y[1] #wartości odpowiadające liczbie zarażonych przechowywane są w drugiej kolumnie
    S_prim=-SIR.r*S*I
    I_prim=SIR.r*S*I-SIR.a*I
    R_prim=SIR.a*I

    return np.array([S_prim,I_prim,R_prim])


def Runge_Kutta4(x0,t0,tk,h):
    
    t=np.arange(t0,tk,h) #wektor czasów od t0 do tk z krokiem h
    N=len(t)             #liczba kroków czasowych

    if type(x0)!=int:
        x=np.zeros((N,len(x0)))  #macierz wynikowa o odpowiednim rozmiarze (dla jednego równania - wektor)
    else: 
        x=np.zeros(N)
        
    x[0]=np.array(x0)  #wpisujemy wartości poczatkowe do pierwszego wiersza, a kolejne wiersze wyliczamy z wzorów:
    
    for n in range(N-1): #Pętla wykona się N-1 razy, ponieważ pierwszy wiersz jest już wpisany
        
        k1=h*SIR(t[n],x[n])
        k2=h*SIR(t[n]+h*0.5,x[n]+0.5*k1)
        k3=h*SIR(t[n]+h*0.5,x[n]+0.5*k2)
        k4=h*SIR(t[n]+h,x[n]+k3)
        x[n+1]=np.array(x[n]+(k1+2.0*k2+2.0*k3+k4)/6)
        
    return t,x

```

## 5. Badanie kiedy wybuchnie epidemia

W przypadku modelu infekcji najciekawsze jest to, jakie są warunki na parametry, aby infekcja zgasła, a przy jakich się rozwinie. Przeprowadzimy doświadczenie, sprawdzające jak będą się zmieniać stany epidemii w zależności od wartości współczynników $a$ i $r$.

W naszym eksperymencie zakładamy, że badaną populacją jest populacja chomików syryjskich, znajdująca się w izolowanym pomieszczeniu. Wszystkie są płci męskiej, więc niemożliwe jest ich rozmnażanie w czasie trwania doświadczenia. Badana grupa liczy 300 osobników, z czego dwadzieścia z nich jest zarażonych ektoparazytozą ($I_0=20$). Żaden nie jest uodporniony na chorobę ($R_0=0$), a pozostałe są podatne na zarażenie ($S_0=280$). Czas trwania doświadczenia to 90 dni.

W chwili $t=0$, mamy:

<h4><center>
$\begin{equation}
\frac{dI}{dt}=-rS_0I_0-aI_0=I_0(rS_0-a)
\end{equation}$.
</center></h4>

Istnieją zatem dwie możliwości:
- Liczba zarażonych będzie początkowo wzrastać, infekcja będzie się rozprzestrzeniać

<h4><center>
$\begin{equation}
\frac{dI}{dt}>0  \Leftrightarrow rS_0-a>0, 
\end{equation}$
 </center></h4>
<h4><center>
$\begin{equation}
~czyli~ S_0> \frac{a}{r}
\end{equation}$
</center></h4>



```python
from ipynb.fs.full.kody import SIR
from ipynb.fs.full.kody import Runge_Kutta4
import matplotlib.pyplot as plt

SIR.a=0.05
SIR.r=0.0009

t,x=Runge_Kutta4([280.0,20.0,0.0],0.0,90.0,0.01)

plt.plot(t,x[:,0],label="S (Suspectible)") #Wykres ilości osobników podatnych na zarażenie od czasu
plt.plot(t,x[:,1],label="I (Infected)")    #Wykres ilości osobnikó zarażonych od czasu
plt.plot(t,x[:,2],label="R (Recovered)")   #Wykres ilości osobników ozdrowiałych od czasu
plt.legend()
plt.xlabel("czas w dniach")
plt.ylabel("ilość osobników")
plt.title("Symulacja dla $ S_0> \dfrac{a}{r} $")
plt.tight_layout()
```

Wniosek - epidemia wybuchnie.

- Liczba zarażonych będzie maleć, infekcja będzie zanikać.
<h4><center>
$\begin{equation}
\frac{dI}{dt}<0  \Leftrightarrow rS_0-a<0,
\end{equation}$
    </center></h4>
<h4><center>
$\begin{equation}
~czyli~ S_0< \frac{a}{r}
\end{equation}$
</center></h4>

```python
from ipynb.fs.full.kody import SIR
from ipynb.fs.full.kody import Runge_Kutta4
import matplotlib.pyplot as plt

SIR.a=0.5
SIR.r=0.0008

t,x=Runge_Kutta4([280.0,20.0,0.0],0.0,90.0,0.01)

plt.plot(t,x[:,0],label="S (Suspectible)")
plt.plot(t,x[:,1],label="I (Infected)")
plt.plot(t,x[:,2],label="R (Recovered)")
plt.legend()
plt.xlabel("czas w dniach")
plt.ylabel("ilość osobników")
plt.title("Symulacja dla $ S_0 < \dfrac{a}{r} $")
plt.tight_layout()
```

Wniosek - epidemia nie wybuchnie.

W zagadnieniu tym pojawia się zjawisko progowe. Parametr $\begin{equation}
p:= \frac{a}{r}
\end{equation}$
 nazywany jest względnym współczynnikiem usuwania zarażonych osobników z ogółu populacji. [5]


Z wykonanych obliczeń wnioskujemy, że epidemia wybucha, jeśli $S_{0}>p$. Zatem pojawienie się epidemii zależy od tego, jak duża jest populacja osobników podatnych na zarażenie w chwili początkowej $\left ( t=0 \right)$.

## 5. Maksymalna ilość zachorowań

Z układu równań (4) wiemy, że  $\frac{dI}{dt}=0 $, gdy  $S=\frac{a}{r}$ lub $I =0$. Badaniem maksymalnej liczby zachorowań będziemy się zajmować w przypadku wybuchu epidemii, czyli gdy liczba zarażonych będzie początkowo wzrastać oraz będzie przynajmniej jeden osobnik w grupie I, który będzie zarażał innych. Stąd odrzucamy przypadek $I =0$ i oczekujemy, że w punkcie $S=\frac{a}{r}$, będzie maksimum lokalne. [6]

Aby znaleźć wzór na $I$, podzielimy przez siebie dwa pierwsze równania z układu (1) (dla łatwiejszych rachunków podzielimy $\frac{dI}{dS}$). Otrzymujemy:


$$
\begin{equation}
\frac{dI}{dS} = \frac{rSI-aI}{-rSI} = -1 + \frac{a}{rS} = -1 + \frac{p}{S}.
\end{equation}
$$

Korzystając z metody przez rozdzielenie zmiennych, otrzymujemy:
   $$
\begin{equation}
dI = \left(-1 + \frac{p}{S}\right)dS.
\end{equation}
$$
Teraz scałkujemy obie strony równania na określonych przedziałach:

$$
\begin{equation}
\int_{I_{0}}^{I}{dI} = \int_{S_{0}}^{S}{\left( -1 + \frac{p}{S}\right)}dS.
\end{equation}
$$

Wykonując powyższe działanie, otrzymujemy:

$$
\begin{equation}
I - I_{0} = - S + S_{0} + p\ln{S} - p\ln{S_{0}}.
\end{equation}
$$

Korzystając ze wzoru (2) oraz z założenia $R_{0}=0$, możemy wykonać podstawienie $N=S_{0}+I_{0}$.
Przyjmujemy także $S=\frac{a}{r}=p$.

$$
\begin{equation}
I  =  N-S_{0}-p+S_{0}+p\ln{p}-p\ln{S_{0}}.
\end{equation}
$$

Ostateczny wzór na maksymalną liczbę zachorowań wygląda następująco:


$$
\begin{equation}
I_{max}  =  N-p+p\ln{\frac{p}{S_{0}}}.
\tag{7}
\end{equation}
$$

Zastosujemy wzór (7) do danych z naszego doświadczenia, aby sprawdzić, ile maksymalnie chomików może zostać zarażonych oraz kiedy to nastąpi. 

```python
from ipynb.fs.full.kody import SIR
from ipynb.fs.full.kody import Runge_Kutta4
import matplotlib.pyplot as plt
import math
import numpy as np

def I_max(a,r,S0,N):
    I=N-a/r +(a/r)*math.log((a/r)/S0)
    print("Maksymalna liczba zachorowań: ", math.floor(I))
    print("Moment w którym zostane osiągnięta:", math.ceil(a/r))
    return math.floor(I)

SIR.a=0.05
SIR.r=0.0009

imax=I_max(SIR.a,SIR.r,280,300)
i = imax * np.ones(9000)        #Tworzymy wektor o długości ilości kroków czasowych i wartości imax

t,x=Runge_Kutta4([280.0,20.0,0.0],0.0,90.0,0.01)

plt.plot(t,x[:,0],label="S")
plt.plot(t,x[:,1],label="I")
plt.plot(t,x[:,2],label="R")
plt.plot(t,i,label=("$I_{max}$"))   #RWykres prostej Imax
plt.legend(loc='best')
plt.xlabel("czas w dniach")
plt.ylabel("ilość osób")
plt.title("Symulacja dla $ S_0> \dfrac{a}{r} $")
plt.tight_layout()
```

W badanym przypadku $I_{max} = 154$. Epidemia osiągnie swoje apogeum w 56 dniu.

## 6. Portret fazowy

Portret fazowy to rodzina trajektorii, przedstawiających zachowanie obiektu obserwowane przy stałym wymuszeniu, ale dla różnych warunków początkowych, które są wówczas jedyną przyczyną zmian obserwowanych w układzie. Jest to wykres krzywych $\{{S(t),I(t), t \in \mathbb{R}}\}$, gdzie $(S(t),I(t))$ są dowolnymi rozwiązaniami (to znaczy dla każdych warunków początkowych). Każda trajektoria portretu reprezentuje ewolucję stanu obiektu od określonego warunku początkowego.

Wykres przedstawia portret fazowy dwóch pierwszych równań z układu (1) dla sytuacji, w której wybucha epidemia.

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
a = 0.05
r = 0.0009
f, g, S, I = symbols('f g S I')
f = -r*S*I
g = r*S*I - a*I

fl = lambdify((S,I), f)
gl = lambdify((S,I), g)

Y, X = np.mgrid[0:300, 0:300]

U = fl(X, Y)
V = gl(X, Y)

plt.streamplot(X, Y, U, V)
plt.xlabel("$S$")
plt.ylabel("$I$")
plt.title("Portret fazowy dla sytuacji, w której epidemia wybuchnie")
plt.tight_layout()
```

Jest to wykres zależności osobników zainfekowanych od osobników podatnych na chorobę. Strzałki na wykresie pokazują kierunek zmiany rozwiązań kiedy $~t$ rośnie. Kiedy ilość osobników zainfekowanych rośnie, ilość osobników podatnych maleje. Od pewnego momentu, obie wartości zdają się zbiegać do 0. Z wcześniejszej analizy wiemy, że wartość początkowa $S_{0}$ znacząco wpływa na rozwój epidemii. Stąd kiedy jest duża ilość zainfekowanych osobników w danej populacji, widać znaczący wzrost liczby zainfekowanych. Im bardziej liczba osobników podatnych na chorobę będzie oddalona od współczynnika progowego (większa), tym więcej osób zostanie zainfekowanych.

```python
from ipynb.fs.full.kody import SIR
from ipynb.fs.full.kody import Runge_Kutta4
import random
import matplotlib.pyplot as plt

#W pętli losujemy losowe liczby z przedziału [0,300], które będą odpowiadały liczbie osobników podatnych na chorobę i zarażonych
#Liczby te sumują się do 300, by obrazowały populację badaną w doświadczeniu

for i in range (5000):
    S0=random.randint(0,300)
    I0=random.randint(0,300)
    if I0+S0==300:
        t,x=Runge_Kutta4([S0,I0,0.0],0.0,90.0,0.01)
        plt.plot(x[:,0], x[:, 1],'g')

plt.plot([300,0],[0,300], 'b',label="$N=I+S$")
plt.plot([SIR.a/SIR.r,SIR.a/SIR.r],[0,300],'k',label ="$S=p$")

plt.xlabel('$S$')
plt.ylabel('$I$')
plt.title('Przestrzeń fazowa')
plt.legend()
plt.tight_layout()
```

Rozważamy populację liczącą 300 osobników. Zatem krzywe na wykresie fazowym nigdy nie przekroczą prostej $N=S+I$. Co dokładnie widać na załączonym wykresie. Co więcej, prosta przechodząca przez punkt $\frac{a}{r}$, wyznacza próg zachowań krzywych. Populacja zainfekowanych rośnie aż do momentu, gdy populacja osobników podatnych zmaleje do poziomu $p$.
Każdy początkowy punkt po prawej stronie linii prowadzi do epidemii. Każdy punkt początkowy na lewo od linii daje monotonię - zanik liczby zarażonych. Oczywiście wszystkie możliwe stany końcowe znajdują się na osi S, po lewej stronie linii. [7]

Portret fazowy dla sytuacji, w której epidemia nie wybuchła.

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
a = 0.5
r = 0.0008
f, g, S, I = symbols('f g S I')
f = -r*S*I
g = r*S*I - a*I

fl = lambdify((S,I), f)
gl = lambdify((S,I), g)

Y, X = np.mgrid[0:300, 0:300]

U = fl(X, Y)
V = gl(X, Y)

plt.streamplot(X, Y, U, V)
plt.xlabel("$S$")
plt.ylabel("$I$")
plt.title("Portret fazowy dla sytuacji, w której epidemia nie wybuchnie")
plt.tight_layout()
```

Możemy zauawżyć, że w tej sytuacji wykres wygląda inaczej niż w przypadku, w którym epidemia wybuchła. Kształt portretu zdefiniowany jest przez inne trajektorie, a rozwiązania $S$ nie zbiegają już do 0, lecz osiągają wszystkie możliwe wartości na osi OS. Natomiast  $I$ dąży do 0, niezależnie od warunków początkowych.

## 7. Wnioski

Na podstawie przeprowadzonego doświadczenia można dokładnie przeanalizować rozwój ektoparazytozy na grupie badawczej liczącej 300 chomików syryjskich. Warunki początkowe: (280 chomików podatnych na zarażenie, 20 chomików zarażonych, 0 chomików odpornych/ozdrowiałych). Warunki końcowe: (2 chomiki podatne na chorobę, 7 zainfekowanych, 291 ozdrowiałych lub martwych). W 56 dniu eksperymentu była największa liczba osobników chorych, tj. 154 chomiki. 

Na podstawie wykresów stworzonych przy pomocy algorytmu Rungego - Kutty rzędu 4, możemy zauważyć, że w rozważanym modelu infekcja zawsze wygaśnie, może nastąpić to w różnym czasie i z różną maksymalną ilością osobników zarażonych, w zależności od tego jak dużą populację będziemy obserwować oraz jakie są panujące w niej warunki początkowe. Dzieje się tak dlatego, że w modelu SIR nie bierzemy pod uwagę tego, że liczność populacji może ulegać zmianie (tzn. nie uwzględnia on wzrostu liczby osobników na skutek narodzin oraz zmniejszenia na skutek śmierci). W pewnym momencie przeprowadzania doświadczenia znajdziemy się w sytuacji, w której nie ma już osobników podatnych na chorobę, ponieważ wszystkie osobniki uzyskały odporność na skutek przebytej choroby lub nie zostanie nam żaden osobnik zainfekowany zdolny do przekazania patogenu osobnikom podatnym. 

W związku, z czym możemy wysnuć wniosek, że nie jest to idealny model do badania rozwoju epidemii w warunkach naturalnych. Aby uzyskać bardziej wiarygodne wyniki, powinniśmy wziąć pod uwagę możliwość zmian liczności populacji to, że niektóre osobniki w momencie przeprowadzania badania są uodpornione na dany patogen oraz fakt, że różne osobniki charakteryzują się różną podatnością na wirusy. Czynnikiem, który w warunkach naturalnych również mógłby mieć wpływ na to czy infekcja wybuchnie, czy też nie jest rozmieszczenie osobników na obszarze geograficznym oraz to, że nie każdy osobnik ma takie same szanse na kontakt z osobnikami zainfekowanymi.

Pomimo tego, że rozważany przez nas model posiada pewne braki, przeprowadzenie powyższej analizy pokazało nam, jak wiele jesteśmy w stanie dowiedzieć się o rozwoju choroby, posiadając niewielką liczbę danych. To pokazuje, że modele matematyczne są w stanie w bardzo dobry sposób zobrazować rzeczywistość, nawet jeśli rozważamy zjawiska, które pozornie mogą wydawać się nieprzewidywalne.
